**In this exercise, we need to obtain the parameters of the Drude model by fitting the relationship to the experimental data.**

***We will do this for the metals Al, Pt, Ag, Cu, and Au, and we will also extract the experimental data from the refractiveindex library.***


In this section, we will use the **RefractiveIndexMaterial** library to obtain the results from Johnson and Christy (1972) and Rakic (1995)

***Since the returned values for permittivity are complex, in this section, we will plot the experimental results of Johnson and Christy in two separate graphs: one for the real part and another for the imaginary part of the permittivity.***

```python
import numpy as np
import matplotlib.pyplot as plt
from refractiveindex import RefractiveIndexMaterial
def wavelength_to_energy(wavelength_nm):
    h = 4.135667696e-15  # eV·s
    c = 3e8  # m/s
    wavelength_m = wavelength_nm * 1e-9  # m
    frequency = c / wavelength_m  # Hz
    energy_eV = h * frequency  # eV
    return energy_eV
#get element information from refractiveindex - for Au-Cu-Ag-Al-Pt

Au = RefractiveIndexMaterial(shelf='main', book='Au', page='Johnson')
Ag = RefractiveIndexMaterial(shelf='main', book='Ag', page='Johnson')
Pt = RefractiveIndexMaterial(shelf='main', book='Pt', page='Werner')
Cu = RefractiveIndexMaterial(shelf='main', book='Cu', page='Johnson')
Al= RefractiveIndexMaterial(shelf='main', book='Al', page='Rakic')
wavelength_nm = np.linspace(350, 1200, 50)
energies_1 = [wavelength_to_energy(w) for w in wavelength_nm]



epsilon_values_Au = [(Au.get_epsilon(w)) for w in wavelength_nm]
epsilon_v_real_Au_exp  = [np.real(w) for w in epsilon_values_Au]
epsilon_v_img_Au_exp = [np.imag(w) for w in epsilon_values_Au]

epsilon_values_Ag = [(Ag.get_epsilon(w)) for w in wavelength_nm]
epsilon_v_real_Ag_exp  = [np.real(w) for w in epsilon_values_Ag]
epsilon_v_img_Ag_exp = [np.imag(w) for w in epsilon_values_Ag]

epsilon_values_Al = [(Al.get_epsilon(w)) for w in wavelength_nm]
epsilon_v_real_Al_exp  = [np.real(w) for w in epsilon_values_Al]
epsilon_v_img_Al_exp = [np.imag(w) for w in epsilon_values_Al]

epsilon_values_Cu = [(Cu.get_epsilon(w)) for w in wavelength_nm]
epsilon_v_real_Cu_exp  = [np.real(w) for w in epsilon_values_Cu]
epsilon_v_img_Cu_exp = [np.imag(w) for w in epsilon_values_Cu]

epsilon_values_Pt = [(Pt.get_epsilon(w)) for w in wavelength_nm]
epsilon_v_real_Pt_exp  = [np.real(w) for w in epsilon_values_Pt]
epsilon_v_img_Pt_exp = [np.imag(w) for w in epsilon_values_Pt]

plt.figure(figsize=(16, 24))


#Au-------------
plt.subplot(5, 2, 1)
plt.plot(energies_1, epsilon_v_real_Au_exp, label='(Au)', color='blue',marker = 'o',linestyle=' ')
plt.title('Real part of Permittivity - experiment:Johnson and Christy 1972 ')
plt.xlabel('Energy (eV)')
plt.ylabel('εr real')
plt.grid()
plt.legend()

plt.subplot(5, 2, 2)
plt.plot(energies_1, epsilon_v_img_Au_exp, label='(Au)', color='red',marker = 'o',linestyle=' ')
plt.title('imaginary part of Permittivity - experiment:Johnson and Christy 1972')
plt.xlabel('Energy (eV)')
plt.ylabel('εr imag')
plt.legend()
plt.grid()

#Ag-------------
plt.subplot(5, 2, 3)
plt.plot(energies_1, epsilon_v_real_Ag_exp, label='(Ag)', color='blue',marker = 'o',linestyle=' ')
plt.title('Real part of Permittivity - experiment:Johnson and Christy 1972 ')
plt.xlabel('Energy (eV)')
plt.ylabel('εr real')
plt.grid()
plt.legend()

plt.subplot(5, 2, 4)
plt.plot(energies_1, epsilon_v_img_Ag_exp, label='(Ag)', color='red',marker = 'o',linestyle=' ')
plt.title('imaginary part of Permittivity - experiment:Johnson and Christy 1972')
plt.xlabel('Energy (eV)')
plt.ylabel('εr imag')
plt.legend()
plt.grid()

#Al-------------
plt.subplot(5, 2, 5)
plt.plot(energies_1, epsilon_v_real_Al_exp, label='(Al)', color='blue',marker = 'o',linestyle=' ')
plt.title('Real part of Permittivity - experiment:Rakić 1995 ')
plt.xlabel('Energy (eV)')
plt.ylabel('εr real')
plt.grid()
plt.legend()

plt.subplot(5, 2, 6)
plt.plot(energies_1, epsilon_v_img_Al_exp, label='(Al)', color='red',marker = 'o',linestyle=' ')
plt.title('imaginary part of Permittivity - experiment:Rakić 1995')
plt.xlabel('Energy (eV)')
plt.ylabel('εr imag')
plt.legend()
plt.grid()

#Pt-------------

plt.subplot(5, 2, 7)
plt.plot(energies_1, epsilon_v_real_Pt_exp, label='(Pt)', color='blue',marker = 'o',linestyle=' ')
plt.title('Real part of Permittivity - experiment:Werner Werner-2009')
plt.xlabel('Energy (eV)')
plt.ylabel('εr real')
plt.grid()
plt.legend()

plt.subplot(5, 2, 8)
plt.plot(energies_1, epsilon_v_img_Pt_exp, label='(Pt)', color='red',marker = 'o',linestyle=' ')
plt.title('imaginary part of Permittivity - experiment:Werner-2009')
plt.xlabel('Energy (eV)')
plt.ylabel('εr imag')
plt.legend()
plt.grid()

#Cu-------------
plt.subplot(5, 2, 9)
plt.plot(energies_1, epsilon_v_real_Cu_exp, label='(Cu)', color='blue',marker = 'o',linestyle=' ')
plt.title('Real part of Permittivity - experiment:Johnson and Christy 1972 ')
plt.xlabel('Energy (eV)')
plt.ylabel('εr real')
plt.grid()
plt.legend()

plt.subplot(5, 2, 10)
plt.plot(energies_1, epsilon_v_img_Cu_exp, label='(Cu)', color='red',marker = 'o',linestyle=' ')
plt.title('imaginary part of Permittivity - experiment:Johnson and Christy 1972')
plt.xlabel('Energy (eV)')
plt.ylabel('εr imag')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def wavelength_to_energy(wavelength_nm):
    h = 4.135667696e-15  # eV·s
    c = 3e8  # m/s
    wavelength_m = wavelength_nm * 1e-9  # m
    frequency = c / wavelength_m  # Hz
    energy_eV = h * frequency  # eV
    return energy_eV
def epsilon_r(omega,k_betta,w_plasma,gamma):
    fpart=k_betta-((w_plasma**2)/(omega*(omega+((1j)*gamma))))
    return fpart
def wavelength_to_w(wavelength_nm):
    c = 3e8  # m/s
    wavelength_m = wavelength_nm * 1e-9  # m
    frequency = c / wavelength_m  # Hz
    angular_velocity = 2 * np.pi * frequency
    return angular_velocity
wavelength_nm = np.linspace(350, 1200, 100)
```

```python
#parameters of Au
wp=1.38e16
gamma=1.07e14 #inverse of mean free path
k_betta=9.71

angular_velocities = [wavelength_to_w(w) for w in wavelength_nm]
energies = [wavelength_to_energy(w) for w in wavelength_nm]
epsilon_values = [epsilon_r(w,k_betta,wp,gamma) for w in angular_velocities]
epsilon_v_real  = [np.real(w) for w in epsilon_values]
epsilon_v_img = [np.imag(w) for w in epsilon_values]

plt.figure(figsize=(18, 6))
plt.subplot(1, 2, 1)
plt.plot(energies, epsilon_v_real, label='drude model(Gold-Au)', color='red')
plt.plot(energies_1, epsilon_v_real_Au_exp, label='in experiment(Gold-Au)', color='blue',marker = 'o',linestyle=' ')
plt.title('Real part of Permittivity ')
plt.xlabel('Energy (eV)')
plt.ylabel('εr')
plt.grid()
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(energies, epsilon_v_img, label='drude model(Gold-Au)', color='red')
plt.plot(energies_1, epsilon_v_img_Au_exp, label='in experiment(Gold-Au)', color='blue',marker = 'o',linestyle=' ')
plt.title('imaginary part')
plt.xlabel('Energy (eV)')
plt.ylabel('εr')
plt.legend()
plt.grid()
```
